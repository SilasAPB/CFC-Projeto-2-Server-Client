```python
import numpy as np
import matplotlib.pyplot as plt

# Parmetros
fs = 44100  # Frequncia de amostragem em Hz
duracao = 2  # Durao em segundos
t = np.linspace(0, duracao, int(fs * duracao), endpoint=False)  # Vetor de tempo

# Senoides
x1 = 4 * np.sin(2 * np.pi * 3000 * t)
x2 = 20 * np.sin(2 * np.pi * 5000 * t)
x3 = 5 * np.sin(2 * np.pi * 6000 * t)

# Soma das senoides
sinal = x1 + x2 + x3

# Funo para plotar a FFT
def plot_fft(sinal, fs):
    N = len(sinal)
    fft_result = np.fft.fft(sinal)
    freqs = np.fft.fftfreq(N, d=1/fs)
    magnitude = np.abs(fft_result)

    # Manter apenas frequncias positivas
    half_N = N // 2
    freqs = freqs[:half_N]
    magnitude = magnitude[:half_N]

    # Plot do sinal no tempo
    plt.figure(figsize=(12, 6))
    plt.subplot(2, 1, 1)
    plt.plot(t, sinal)
    plt.title("Sinal no Tempo")
    plt.xlabel("Tempo (s)")
    plt.ylabel("Amplitude")
    plt.grid(True)

    # Plot da FFT
    plt.subplot(2, 1, 2)
    plt.stem(freqs, magnitude, basefmt=" ")
    plt.title("Transformada de Fourier (FFT)")
    plt.xlabel("Frequncia (Hz)")
    plt.ylabel("Magnitude")
    plt.grid(True)
    plt.tight_layout()
    plt.show()

# Plotar a FFT do sinal
plot_fft(sinal, fs)
```

```python
# Adicionar rudo ao sinal
ruido = np.random.normal(0, 2, len(sinal))  # Rudo com intensidade 2
sinal_com_ruido = sinal + ruido

# Plotar a FFT do sinal com rudo
plot_fft(sinal_com_ruido, fs)

# Aumentar a intensidade do rudo
ruido_intenso = np.random.normal(0, 100, len(sinal))  # Rudo com intensidade 100
sinal_com_ruido_intenso = sinal + ruido_intenso

# Plotar a FFT do sinal com rudo intenso
plot_fft(sinal_com_ruido_intenso, fs)
```
